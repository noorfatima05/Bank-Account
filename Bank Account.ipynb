```python
class BankAccount ():
    def __init__(self, customer_name, account_num, date_of_opening, balance):
        self.customer_name = customer_name
        self.account_num = account_num
        self.date_of_opening = date_of_opening
        self.balance = balance
    def deposit (self, amount):
        self.balance += amount
        print( f"Rs {amount} is deposit in your account")
    def withdraw (self, amount):
        if amount > self.balance:
            print("insuffisent amount")
        else:
            self.balance -= amount
            print( f"{amount} is withdraw in your account")
    def check_balance(self):
        print(f"this is the balance in your account Rs{self.balance}")
    def customer_details(self):
        print("Customer Name:",self.customer_name)
        print("Account Number:",self.account_num)
        print("Date_of_opening:",self.date_of_opening)
        print("Balance:",self.balance)
              
account1=BankAccount("Zeeshan","E3457",12-4-2023,8000)              
account2=BankAccount("Aqsa","E3458",23-6-2021,2000)
account3=BankAccount("Shani","E3459",1-7-2022,7000)
account4=BankAccount("Nooor","E3460",20-9-2020,6000)  
              
account1.customer_details()
print("-------------------")
account2.customer_details()
print("-------------------")
account3.customer_details()
print("-------------------")
account4.customer_details()
print("-------------------")

print("Deposit Amount")
account1.deposit(3000)
print("-------------------")
account2.deposit(1000)
print("-------------------")
account3.deposit(2000)
print("-------------------")
account4.deposit(4000)
print("-------------------")

print("withdraw amount")
account1.withdraw(4000)
print("check balance")
account1.check_balance()
print("withdraw amount")
account1.withdraw(8000)
print("withdraw amount")
account1.withdraw(7000)
print("check balance")
account1.check_balance()
```
